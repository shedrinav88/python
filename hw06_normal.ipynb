```python
# У каждого ученика есть два Родителя(мама и папа).
# Также в школе преподают Учителя. Один учитель может преподавать 
# в неограниченном кол-ве классов свой определенный предмет. 
# Т.е. Учитель Иванов может преподавать математику у 5А и 6Б,
# но больше математику не может преподавать никто другой.

# Выбранная и заполненная данными структура должна решать следующие задачи:
# 1. Получить полный список всех классов школы
# 2. Получить список всех учеников в указанном классе
#  (каждый ученик отображается в формате "Фамилия И.О.")
# 3. Получить список всех предметов указанного ученика 
#  (Ученик --> Класс --> Учителя --> Предметы) 
# 4. Узнать ФИО родителей указанного ученика
# 5. Получить список всех Учителей, преподающих в указанном классе

class People:
    def __init__(self, name, surname, mid_name): 
        self.name = name
        self.surname = surname
        self.mid_name = mid_name
    def surname_init(self):
        return '{} {}.{}.'.format(self.surname,self.name[0], self.mid_name[0])
    def fio(self):
        return '{} {} {}'.format(self.surname, self.name, self.mid_name)
    
class Parents(People):
    def __init__(self, name, surname, mid_name, child, relationship):
        People.__init__(self, name, surname, mid_name)
        self.child = child
        self.relationship = relationship
class Teacher(People):
    def __init__(self, name, surname, mid_name, grades, subject):
        People.__init__(self, name, surname, mid_name)
        self.grades = grades
        self.subject = subject
    def biglist(self):
        for _ in self.grades:
            return '{}-->{}-->{}'.format(_, self.fio(), self.subject)
        
class Student(People):
    def __init__(self, name, surname, mid_name, mom, dad, grade, subjects):
        People.__init__(self, name, surname, mid_name)
        self.mom = mom
        self.dad = dad
        self.grade = grade
        self.subjects = subjects
        

def class_list(teachers):
    a = []
    for i in teachers:
        a.append(i.grades)
    b = []
    for i in a:
        b = b + i
    print('В ШКОЛЕ ОБУЧАЮТСЯ СЛЕДУЮЩИЕ КЛАССЫ:')
    for i, k in enumerate(set(b), start = 1):
        print('{}) {} класс'.format(i, k))
    
def student_list(students):
    a = input('Список учеников какого класса необходимо посмотреть (формат ввода: 5 A)? ')
    d = []
    for i in students:
        if i.grade == a:          
            d.append(i.surname_init())
    if bool(d) == False:
        return print('Класс введен некорректно, либо не существует!')
    print('Список учеников ' + a + ' класса:')
    for i, k in enumerate(set(d), start = 1):
        return print('{}) {}'.format(i, k))
def teachers_list(teachers):
    a = input('Список преподавателей какого класса необходимо посмотреть (формат ввода: 5 A)? ')
    d = []
    for i in teachers:
        for k in i.grades:
            if k == a:
                d.append(i)
    if bool(d) == False:
        return print('Класс введен некорректно, либо не существует!')
    print('Список преподавателей ' + a + ' класса:')
    for i, k in enumerate(set(d), start = 1):
        print('{}) {}--> {}'.format(i, k.surname_init(), k.subject))
def parents_list(parents):
    a = input('Родителей какого ученика нужно найти (формат ввода: Petrov E.A.)? ')
    for i in parents:
        if i.child == a:
            print('{}: {}'.format(i.relationship, i.fio()))

def student_info(students, teachers):            
    a = input('По какому ученику предоставить информацию (формат ввода: Petrov E.A.)? ')
    for i in students:
        if i.surname_init() == a:
            print('{}--> {} class -->'.format(i.surname_init(), i.grade)) 
            for k in i.subjects:
                for n in teachers:
                    if k == n.subject:
                        print('{:>37} {}--> {}'.format('teacher:', n.surname_init(), n.subject))

def print_menu():
    print('ЗДРАВСТВУЙТЕ! Я ВАШ ПОМОЩНИК! ВОТ ЧТО Я МОГУ:')
    print('[1] Получить полный список всех классов школы')
    print('[2] Получить список всех учеников в указанном классе')
    print('[3] Получить список всех предметов указанного ученика') 
    print('[4] Узнать ФИО родителей указанного ученика')
    print('[5] Получить список всех Учителей, преподающих в указанном классе')
    print('[q] Выход из меню')


parents = [Parents('Alexandr', 'Petrov', 'Ivanovich', 'Petrov E.A.', 'dad'),
        Parents('Anna', 'Petrova', 'Alexeevna', 'Petrov E.A.', 'mom'),
        Parents('Ilya', 'Sviridov', 'Ivanovich', 'Sviridova Y.I.', 'dad'),
        Parents('Kate', 'Sviridova', 'Petrovna', 'Sviridova Y.I.', 'mom'),
        Parents('Anatoliy', 'Galitsky', 'Mihailovich', 'Galitsky I.A.', 'dad'),
        Parents('Irina', 'Shayk', 'Afanasyevna', 'Galitsky I.A.', 'mom')]

teachers = [Teacher('Evgeniy', 'Orlov', 'Ivanovich', ['7 B', '6 G'], 'math'),
            Teacher('Irina', 'Brown', 'Petrovna', ['5 A', '7 B', '6 G',], 'russian language'),
            Teacher('Oleg', 'Petrov', 'Ivanovich', ['7 B'], 'chemistry')]

students = [Student('Egor', 'Petrov', 'Alexandrovich', 'Petrova Anna Alexeevna', 'Petrov Alexandr Ivanovich', '7 B', ['math', 'russian language','chemistry' ]),
            Student('Yulia', 'Sviridova', 'Ilyinichna','Kate Petrovna Sviridova', 'Ilya Ivanovich Sviridov', '5 A', ['russian language']),
            Student('Ivan', 'Galitsky', 'Anatolievich','Irina Afanasyevna Shayk', 'Anatoliy Mihailovich Galitsky', '6 G', ['math', 'russian language'])]

print_menu()    
choose = ''
while True:
    choose = input('Какую информацию из меню необходимо предоставить? ')
    if choose == '1':
        class_list(teachers)
    elif choose == '2':
        student_list(students)
    elif choose == '3':
        student_info(students, teachers)
    elif choose == '4':
        parents_list(parents)
    elif choose == '5':
        teachers_list(teachers)
    elif choose == 'q' or choose == 'Q':
        print('До следующей встречи!')
        break
    else:
        print('Вы ввели некорректную комбинацию! Выберите пункт из меню!')
```
