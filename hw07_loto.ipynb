```python
#создаем класс с магическими методами, чтобы красиво выводить карточки на экран
class nice_str:
    def __init__(self, list):
        self.list = list
    def __iter__(self):
        return list(self)
    def __next__(self):
        return list(self)
    def __str__(self):              
        return '{} {} {} {} {} {} {} {} {}'.format(self.list[0], self.list[1], self.list[2], self.list[3], self.list[4], self.list[5], self.list[6], self.list[7], self.list[8])

import random
#данная функция выбирает 5 случайных чисел (3 раза, для каждой из строк карточки) в диапазоне от 0 до 8
#это будут индексы расположения чисел в строках карточки
def ind_list():
    st1 = []
    while len(st1) < 15:
        st = list(range(0, 9))
        for i in range(5):
                choose = random.choice(st)
                st1.append(choose)
                st.remove(choose)
    return [sorted(st1[0:5]), sorted(st1[5:10]), sorted(st1[10:15])]

#данная функция выбирает 15 уникальных чисел из диапазона 1 - 90, после разбивает их на 3 группы по 5 чсел в отсортированном виде
def num_list():
    st = list(range(1, 91))
    st1 = []
    for i in range(15):
            choose = random.choice(st)
            st1.append(choose)
            st.remove(choose)
    return [sorted(st1[0:5]), sorted(st1[5:10]), sorted(st1[10:15])]
#данная функция создает карточку для игры
def card_maker():
    my_card = [[' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
              [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
              [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']]
    u = 0
    a = ind_list()
    s = num_list()
    for i in my_card:
        i[a[u][0]] = s[u][0]
        i[a[u][1]] = s[u][1]
        i[a[u][2]] = s[u][2]
        i[a[u][3]] = s[u][3]
        i[a[u][4]] = s[u][4]
        u += 1
    return my_card

player = card_maker()
player1 = nice_str(player[0])
player2 = nice_str(player[1])
player3 = nice_str(player[2])
computer = card_maker()
computer1 = nice_str(computer[0])
computer2 = nice_str(computer[1])
computer3 = nice_str(computer[2])
v = list(range(1, 91))
t = []
step = 1
while True:
    try:
        for l in range(1):
            choose = random.choice(v)
            t.append(choose)
            v.remove(choose)
    except IndexError:
        StopIteration
        print('Бочонки закончились')
    print('------ХОД ' + str(step) + ' ------')    
    print('                             ')
    print('Новый бочонок: ' + str(choose) + ' (осталось ' + str(len(v)) + ')')
    print('----Ваша карточка----')
    print(player1)
    print(player2)
    print(player3)
    print('-------------------------')
    print('                             ')
    print('--Карточка компьютера---')
    print(computer1)
    print(computer2)
    print(computer3)
    print('-------------------------')
    print('                             ')
    answer = ''
    while True:
        answer = input('Зачеркнуть число? y/n ')
        print('                             ')
        if answer == 'y' or answer == 'Y' or answer == 'n' or answer == 'N':
            break
        else:
            print('Вы ввели некорректную комбинацию, попробуйте снова!')
                
    if answer == 'y' or answer == 'Y':
        if choose in player[0]:
            player[0][player[0].index(choose)] = '-'
        elif choose in player[1]:
            player[1][player[1].index(choose)] = '-'
        elif choose in player[2]:
            player[2][player[2].index(choose)] = '-'
        else:
            print('Числа не было в Вашей карточке! Вы проиграли!')
            break            
    elif answer == 'n' or answer == 'N':
        if choose in player[0]:
            print('Число было в Вашей карточке! Вы проиграли!')
            break
        elif choose in player[1]:
            print('Число было в Вашей карточке! Вы проиграли!')
            break
        elif choose in player[2]:
            print('Число было в Вашей карточке! Вы проиграли!')
            break            
    if choose in computer[0]:
            computer[0][computer[0].index(choose)] = '-'
    elif choose in computer[1]:
            computer[1][computer[1].index(choose)] = '-'
    elif choose in computer[2]:
            computer[2][computer[2].index(choose)] = '-'
    if player[0].count('-') == 5 and player[1].count('-') == 5 and player[2].count('-') == 5:
        print('Поздравляем! Вы победили!')
        break
    elif computer[0].count('-') == 5 and computer[1].count('-') == 5 and computer[2].count('-') == 5:
        print('В этой суровой схватке победу одержал компьютер!')
        break
    step += 1
```
